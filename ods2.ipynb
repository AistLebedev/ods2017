```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("Video_Games_Sales_as_at_22_Dec_2016.csv")
df.info()
```

```python
df
```

```python
df = df.dropna()
print(df.shape)
```

```python
df['User_Score'] = df.User_Score.astype('float64')
df['Year_of_Release'] = df.Year_of_Release.astype('int64')
df['User_Count'] = df.User_Count.astype('int64')
df['Critic_Count'] = df.Critic_Count.astype('int64')
```

```python
df.info()
```

```python
useful_cols = ['Name', 'Platform', 'Year_of_Release', 'Genre', 
               'Global_Sales', 'Critic_Score', 'Critic_Count',
               'User_Score', 'User_Count', 'Rating'
              ]
df[useful_cols].head()
```

```python
sales_df = df[[x for x in df.columns if 'Sales' in x] + ['Year_of_Release']]
sales_df.groupby('Year_of_Release').sum().plot()
```

```python
sales_df.groupby('Year_of_Release').sum().plot(kind='bar', rot=45)
```

**Seaborn**

```python
cols = ['Global_Sales', 'Critic_Score', 'Critic_Count', 'User_Score', 'User_Count']
sns_plot = sns.pairplot(df[cols])
sns_plot.savefig('pairplot.png')
```

```python
sns.distplot(df.Critic_Score)
```

```python
top_platforms = df.Platform.value_counts().sort_values(ascending = False).head(5).index.values
sns.boxplot(y="Platform", x="Critic_Score", data=df[df.Platform.isin(top_platforms)], orient="h")
```

```python
platform_genre_sales = df.pivot_table(
                        index='Platform', 
                        columns='Genre', 
                        values='Global_Sales', 
                        aggfunc=sum).fillna(0).applymap(float)
sns.heatmap(platform_genre_sales, annot=True, fmt=".1f", linewidths=.5)
```

```python
data = pd.read_csv("https://raw.githubusercontent.com/Yorko/mlcourse.ai/master/data/mlbootcamp5_train.csv", sep=";")
```

```python
data
```

```python
data.info()
```

```python
for c in data.columns:
    n = data[c].nunique()
    print(c)
    if n <= 3:
        print(n, sorted(data[c].value_counts().to_dict().items()))
    else:
        print(n)
    print(10 * "-")
```

<h4>1.1 Сколько женщин и мужчин в датасете, зная что мужчины в среднем выше<h4>

<h4>2.1 Наибольшую корреляцию Пирсона с фичей gender<h4>

```python
df = data.copy()

corr = data.corr(method='pearson')


f, ax = plt.subplots(figsize=(12, 9))

sns.heatmap(corr, vmax=1, center=1, annot=True, fmt='.1f',
            square=True, linewidths=.5, cbar_kws={"shrink": .5});
```

<h4>2.2violinplot для роста и пола<h4>

```python
df_melt = pd.melt(frame=data, value_vars=['height'], id_vars=['gender'])
```

```python
df_melt
```

```python
plt.figure(figsize=(12, 10))
ax = sns.violinplot(
    x='variable', 
    y='value', 
    hue='gender', 
    palette="muted", 
    split=True, 
    data=df_melt, 
    scale='count',
    scale_hue=False
)
```

<h4>2.3.1 Ранговая корреляция Спирмена<h4>

```python
corr = df[['id', 'age', 'height', 'weight', 
           'ap_hi', 'ap_lo', 'cholesterol', 
           'gluc']].corr(method='spearman')
f, ax = plt.subplots(figsize=(12, 10))
sns.heatmap(corr, vmax=1, center=0, annot=True, fmt='.2f',
            square=True, linewidths=.5, cbar_kws={"shrink": .5});
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv("https://raw.githubusercontent.com/Yorko/mlcourse.ai/master/data/adult.data.csv")
```

```python
data
```

```python
data["sex"].value_counts() #1
```

```python
#data[data["native-country"] == "Germany"].value_counts() #3
data["native-country"].value_counts(normalize=1)
```

```python
data[data["salary"] == "<=50K"]["age"].describe() #4-5
```

```python
data[data["salary"] == ">50K"]['education'].value_counts() #6
```

```python
data[data["race"] == "Amer-Indian-Eskimo"]['age'].sort_values() #7
```

```python
print(data[data["marital-status"] == "Married-civ-spouse"]['salary'].value_counts(normalize=1))
print(data[data["marital-status"] == "Married-spouse-absent"]['salary'].value_counts(normalize=1))
print(data[data["marital-status"] == "Married-AF-spouse"]['salary'].value_counts(normalize=1)) #8
```

```python
data[data['hours-per-week'] == 99]['salary'].value_counts(normalize=1)#9
```

```python
print(data[(data["salary"] == ">50K") & (data["native-country"] == "Japan")]['hours-per-week'].describe())
print(data[(data["salary"] == "<=50K") & (data["native-country"] == "Japan")]['hours-per-week'].describe()) #10
```

```python
pd.crosstab(data['native-country'], data['salary'], 
           values=data['hours-per-week'], aggfunc=np.mean).T
```
